### Ques on Group by

```python
# Find the top 3 genre by total earning 
x = movies.groupby('Genre')['Gross'].sum().sort_values(ascending=False)
x.head(3)
```

```python
# Find the genre with highest avg IMBD rating 

movies.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False).head(1)
```

```python
# Find the number of movies done by each actor

movies['Star1'].value_counts()

movies.groupby('Star1')['Series_Title'].count().sort_values(ascending=False)
```

### agg method 

analise data column wise while applying different operations

```python
# agg method  --> put multiple method at a time 
genres.agg(
    {
        'Runtime':'mean',
        'IMDB_Rating':'mean',
        'No_of_Votes':['sum','min'],
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

```python
#split(apply) combine 
#apply --> builtin function 

genres.apply(min)
```

```python
# Find number of movies starting wiht A for each group 
def foo(group):
    return group['Series_Title'].str.startswith('A').sum()
   
```

```python
genres.apply(foo)
```

```python
#groupby on multipple column 

duo = movies.groupby(['Director','Star1'])
duo.size()
```

```python
# find the best (in term of metascore(avg)) actor -> genre combo
movies.groupby(['Star1','Genre'])['Metascore'].mean().reset_index().sort_values('Metascore',ascending=False).head(1)
```

```python
# Find the top 10 batsman in terms of runs

ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# Find the batsman with max no of sixes 

six = ipl[ipl['batsman_runs'] == 6]

six.groupby('batsman')['batsman'].count().sort_values(ascending=False)
```

```python
# find batsman with most number of 4 and 6 in last 5 overs
x = ipl[ipl['over'] > 15]
z = x[(x['batsman_runs'] == 4) |  (x['batsman_runs'] == 6)]  
z.groupby('batsman')['batsman'].count().sort_values(ascending=False).head()
```
