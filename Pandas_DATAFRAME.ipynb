```python
# rank series

batsman = pd.read_csv("batsman_runs_ipl.csv")
batsman.head(5)
```

### Dropna and fillna

```python
school.dropna()   #remove all na value default select(how="any")
school.dropna(how="all")
```

```python
# remove row in which class and branch has NaN value
school.dropna(subset=['Branch'])
```

```python
# use fill na to fill missing name with unknown

school["Name"].fillna('Unknown')
```

```python
# forward filling and backward filling

school['Name'].fillna(method='bfill')
```

```python
# drop()  ( remove prticular columns )
school.drop(columns=['Branch','Roll no'])
```

```python
# drop for row
school.set_index('Name').drop(index=['Anas','Abu'])
```
